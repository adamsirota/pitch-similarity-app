```python
import pybaseball as pyb
startdate = '2024-03-28'
enddate = '2024-11-01'
savant_data = pyb.statcast(start_dt=startdate, end_dt=enddate)
savant_data = savant_data[savant_data['game_type']=='R']
savant_data.head()
```

```python
import pandas as pd
old_df = pd.DataFrame(savant_data)
for col in old_df.columns:
    print(col)
```

```python
selected_columns = ['player_name', 'pitch_type', 'release_speed', 'release_pos_x', 'release_pos_z', 'pitcher', 'pfx_x', 'pfx_z', 'release_spin_rate', 'release_extension']
df = old_df[selected_columns]
df.head(50)
```

```python
import pandas as pd
df['pfx_z'] = df['pfx_z'] * 12
df['pfx_x'] = df['pfx_x'] * 12
grouped_df = df.groupby(['player_name', 'pitch_type']).mean().reset_index()
grouped_df.head()
```

```python
import numpy as np
input_pitch = {
    'release_speed': 92.8,
    'pfx_x': 8.9,
    'pfx_z': 16.3,
    'release_spin_rate': 2254,
    'release_extension': 6.6
}
def calculate_distance(row, input_pitch):
    return np.sqrt(
        (row['release_speed'] - input_pitch['release_speed']) ** 2 +
        (row['pfx_x'] - input_pitch['pfx_x']) ** 2 +
        (row['pfx_z'] - input_pitch['pfx_z']) ** 2 +
        (row['release_spin_rate'] - input_pitch['release_spin_rate']) ** 2 +
        (row['release_extension'] - input_pitch['release_extension']) ** 2
    )
grouped_df['distance'] = grouped_df.apply(calculate_distance, input_pitch=input_pitch, axis=1)
similar_pitches = grouped_df.sort_values('distance').reset_index(drop=True)
similar_pitches.head()
```

```python
grouped_df.to_csv('grouped_picthes.csv', index =False)
```
